```python
# Построить столбиковую диаграмму по возрасту и соответствующему их количеству
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn

import pandas as pd
data = pd.DataFrame
data = pd.read_csv('titanic_train.csv')
data = data.sort_values(by='age')
groupby_age = data.groupby('age').size() 
groupby_age.plot.bar(title = 'Distribution of Age', figsize = (15,4)) 
plt.show()
```

```python
# Построить парную диаграмму между количеством пассажиров разделенных по половому признаку и количеству пассажиров разделенных по возрасту
data1 = data.groupby(['age', 'sex']).size().unstack('sex').fillna(0)
seaborn.pairplot(data1)
plt.show()
```
